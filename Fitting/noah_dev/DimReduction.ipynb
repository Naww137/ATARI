```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Generate example data
X, y = make_classification(n_samples=100, n_features=5, n_informative=3, n_classes=2, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_std)

# Plot the original and t-SNE-transformed data
plt.figure(figsize=(12, 4))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(X_std[y == 0, 0], X_std[y == 0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_std[y == 1, 0], X_std[y == 1, 1], color='blue', marker='o', alpha=0.5)
plt.title('Original Data')

# t-SNE-transformed data
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], color='blue', marker='o', alpha=0.5)
plt.title('t-SNE Transformed Data')

plt.tight_layout()
plt.show()
```
